## Visualisasi Heatmap

```python
plt.figure(figsize=(10,8))
sns.heatmap(dataframe.corr(),annot=True,linewidths=0.6,fmt=".2f",cmap="flare")
plt.show()
```

#### Kesimpulan : Bahwa korelasi antara age dan stroke cukup tinggi atau kuat , yaitu sekitar 0.25 sehingga warnanya warna kotaknya agak gelap, dan nilai korelasinya sebesar 0.25. Dan dapat disimpulkan bahwa age sangat mempengaruhi terjadinya seseorang Brain Stroke atau stroke.

## Visualisasi Pie Chart

```python
dfSmokingCounts = dataframe['smoking_status'].where(dataframe['stroke'] == 1).value_counts()
dfSmokingCounts
```

```python
plt.figure(figsize=(15, 5))
sns.set_theme()
plt.pie(dfSmokingCounts, labels=dfSmokingCounts.index, autopct='%.2f%%', shadow=True, startangle=90, explode=[0, 0.1, 0, 0])
plt.title("Persentase Status perokok yang mengalami Brain Stroke", fontsize=20, fontweight="bold")
plt.show()
```

#### Kesimpulan: Berdasarkan Pie Chart diatas, dapat kita lihat bahwa orang yang tidak merokoklah yang paling banyak terkena penyakit Brain Stroke atau stroke, sebanyak 35.89% dari data diatas.

## Visualisasi Histogram

##### Memilih atau ingin menampilkan baris yang pada nilai kolom "stroke"nya == 1 atau terkena Brain Stroke atau Stroke.
