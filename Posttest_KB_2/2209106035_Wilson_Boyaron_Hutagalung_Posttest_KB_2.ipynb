### Menampilkan informasi jumlah record, jumlah attribute, nama attribute, jumlah attribute dengan tipe angka dan memasukkan informasi tersebut dalam dictionary dengan key sesuai nama informasinya.

```python
Dataset_Information = {}

Dataset_Information['Jumlah Record'] = DataFrame.shape[0]
Dataset_Information['Jumlah Attribute'] = DataFrame.shape[1]
Dataset_Information['Nama Attribute'] = DataFrame.columns.tolist()
Dataset_Information['Jumlah Attribute Dengan Tipe Angka'] = DataFrame.select_dtypes(include=['int64', 'float64']).shape[1]

dictKeys = list(Dataset_Information.keys())

table = prtl.PrettyTable()
table.add_column(dictKeys[0], [Dataset_Information['Jumlah Record']])
table.add_column(dictKeys[1], [Dataset_Information['Jumlah Attribute']])
table.add_column(dictKeys[2], ['\n'.join(Dataset_Information['Nama Attribute'])])
table.add_column(dictKeys[3], [Dataset_Information['Jumlah Attribute Dengan Tipe Angka']])
print(table)

```
