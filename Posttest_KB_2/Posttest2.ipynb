### Melakukan Import Library Pandas dan Prettytable dan Import DataSet yang dipakai.

```python
import pandas as pd
import prettytable as prtl
DataFrame = pd.read_csv("Alpha_bank.csv")
```

### Menampilkan rata-rata, standar deviasi, nilai minimum, nilai maksimum, Q1, Q2, dan Q3 dari Data yang bertipe data Numerik(angka).

### Menampilkan 10 Record pertama dari DataSet dengan atribut dengan tipe data Numeric(angka).

```python
DataFrame.select_dtypes(include=['int64', 'float64']).head(10)
```

### Menampilkan informasi jumlah record, jumlah attribute, nama attribute, jumlah attribute dengan tipe angka dan memasukkan informasi tersebut dalam dictionary dengan key sesuai nama informasinya.

```python
datasetInformation = {}

datasetInformation['Jumlah Record'] = DataFrame.shape[0]
datasetInformation['Jumlah Attribute'] = DataFrame.shape[1]
datasetInformation['Nama Attribute'] = DataFrame.columns.tolist()
datasetInformation['Jumlah Attribute Dengan Tipe Angka'] = DataFrame.select_dtypes(include=['int64', 'float64']).shape[1]

dictKeys = list(datasetInformation.keys())

table = prtl.PrettyTable()
table.add_column(dictKeys[0], [datasetInformation['Jumlah Record']])
table.add_column(dictKeys[1], [datasetInformation['Jumlah Attribute']])
table.add_column(dictKeys[2], ['\n'.join(datasetInformation['Nama Attribute'])])
table.add_column(dictKeys[3], [datasetInformation['Jumlah Attribute Dengan Tipe Angka']])
print(table)

```
